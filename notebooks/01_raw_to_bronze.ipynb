```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("DataLake") \
    .config(
        "spark.hadoop.hive.metastore.uris",
        "thrift://docker-hive-hive-metastore-1:9083"
    ) \
    .enableHiveSupport() \
    .getOrCreate()
```

```python
raw_path = "hdfs://docker-hive-namenode-1:8020/data-lake/raw/transaksi/transaksi_raw.csv"

raw_df = spark.read.csv(
    raw_path,
    header=True,
    inferSchema=True
)

raw_df.show()
```

```python
raw_df.printSchema()
```

```python
bronze_path = "hdfs://docker-hive-namenode-1:8020/data-lake/bronze/transaksi"

raw_df.write \
    .mode("overwrite") \
    .parquet(bronze_path)
```
